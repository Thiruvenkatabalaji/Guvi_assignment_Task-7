```python
def standardize_column(df):
    for col in df.columns:
        if col != "Type of glass":
            df[col] = (df[col] - df[col].mean())/df[col].std()
    return df      
```

```python
df_test=standardize_column(df_test)
df_test.head()
```

```python
df_train = standardize_column(df_train)
df_train.head()
```
